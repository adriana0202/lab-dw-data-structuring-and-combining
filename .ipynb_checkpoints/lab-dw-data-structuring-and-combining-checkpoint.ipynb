```python
import pandas as pd
import numpy as np
url1 = 'https://raw.githubusercontent.com/data-bootcamp-v4/data/main/marketing_customer_analysis_clean.csv'
url2='https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file2.csv'
url3='https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file3.csv'


df1 = pd.read_csv(url1)
df2 =pd.read_csv(url2)
df3 =pd.read_csv(url3)
```

```python
df1.head()
```

```python
df1.columns
```

```python
df1=df1.drop('unnamed:_0', axis=1)
df1
```

```python
#match df2 and df3 with df1
def clean_my_columns_titles(df):
    df.columns = df.columns.str.replace(" ","_").str.lower().str.strip()
    
clean_my_columns_titles(df2)
clean_my_columns_titles(df3)
```

```python
df2.rename(columns= {'st': 'state'}, inplace=True)

df2.number_of_open_complaints=df2.number_of_open_complaints.str.split('/').str[1]

df2.customer_lifetime_value=df2.customer_lifetime_value.str.strip('%')
```

```python
df=pd.concat([df1, df2,df3],ignore_index=True)
df.tail(20)
```

# Challenge 2: Structuring Data

In this challenge, we will continue to work with customer data from an insurance company, but we will use a dataset with more columns, called marketing_customer_analysis.csv, which can be found at the following link:

https://raw.githubusercontent.com/data-bootcamp-v4/data/main/marketing_customer_analysis_clean.csv

This dataset contains information such as customer demographics, policy details, vehicle information, and the customer's response to the last marketing campaign. Our goal is to explore and analyze this data by performing data cleaning, formatting, and structuring.

```python
import pandas as pd
import numpy as np
url = 'https://raw.githubusercontent.com/data-bootcamp-v4/data/main/marketing_customer_analysis_clean.csv'
df = pd.read_csv(url)
```

```python
df.monthly_premium_auto.describe()
```

1. You work at the marketing department and you want to know which sales channel brought the most sales in terms of total revenue. Using pivot, create a summary table showing the total revenue for each sales channel (branch, call center, web, and mail).
Round the total revenue to 2 decimal points.  Analyze the resulting table to draw insights.

```python
# creating the summary table: 

summary_table = df.pivot_table(index='sales_channel', values=['monthly_premium_auto'], aggfunc='sum')
summary_table
```

The sales channel drawing the most premium revenue is the Agent channel. Customers do not seem to be used to use their website to get new policies.

2. Create a pivot table that shows the average customer lifetime value per gender and education level. Analyze the resulting table to draw insights.

```python
cust_ltv = df.pivot_table(index='education', columns='gender', values=['customer_lifetime_value'], aggfunc='mean').round(2)
cust_ltv.stack()
```

The policyholders with the best average customer lifetime value across both Make and Females are the ones with Master, High School or below education, while for the rest type of education policyholders are Males except for the Bachelor education where Females lead.
